# Problem Statement : Analyze the given input table and come  with output as shown

```python
# importing required libraries
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# creating dataframe
data=pd.DataFrame()
data['store_id']=[1,1,2,2,2,3]
data['product_1']=['Apple - IPhone','Apple - AirPods','Apple_IPhone','Google Pixel','Sony: Camera','samsung - Galaxy Phone']
data['product_2']=['Apple - MacBook Pro','Samsung - Galaxy Phone','Apple: Phone',' apple: Laptop','Apple Vision Pro','mapple MacBook Pro']
```

```python
data
```

```python
data['product1']=0    # creating new column
data['product2']=0    # creating new column
data.loc[data['product_1'].str.lstrip().str.lower().str.startswith('apple'),['product1']]=1  # applying conditions
data.loc[data['product_2'].str.lstrip().str.lower().str.startswith('apple'),['product2']]=1  # applying conditions
```

```python
df1=pd.DataFrame(data.groupby('store_id')['product1'].sum()).reset_index()     
df2=pd.DataFrame(data.groupby('store_id')['product2'].sum()).reset_index()
```

```python
pd.merge(df1,df2) # final output
```
